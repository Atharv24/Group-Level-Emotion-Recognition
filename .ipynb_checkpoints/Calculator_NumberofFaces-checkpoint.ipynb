# Import Modules

```python
from __future__ import print_function, division

import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from PIL import Image
import numpy as np
import torchvision
from torchvision import datasets, models, transforms
import time
import os
import copy

import matplotlib.pyplot as plt
plt.ion()
```

# Initialize Lists

```python
faces_data = []
faces_data_val = []
```

# Load Train and Val Dataset

```python
neg_train = sorted(os.listdir('Dataset/emotiw/train/'+'Negative/'))
neu_train = sorted(os.listdir('Dataset/emotiw/train/'+'Neutral/'))
pos_train = sorted(os.listdir('Dataset/emotiw/train/'+'Positive/'))

neg_val = sorted(os.listdir('Dataset/emotiw/val/' + 'Negative/'))
neu_val = sorted(os.listdir('Dataset/emotiw/val/' + 'Neutral/'))
pos_val = sorted(os.listdir('Dataset/emotiw/val/' + 'Positive/'))
```

```python
neg_train = neg_train[1:]
pos_val = pos_val[1:]
neu_val = neu_val[1:]
neg_val = neg_val[1:]
```

```python
train_filelist = neg_train + neu_train + pos_train
val_filelist = neg_val + neu_val + pos_val
print(len(train_filelist),len(val_filelist))
```

# Make List of Number of Faces in Train Images

```python
for i in (neg_train):
    file = i
    features = np.load('Dataset/FaceFeatures/train/Negative/'+file[:-3]+'npz')['a']
    faces_data.append(features.shape[0])
    print(i)
```

```python
for i in (neu_train):
    file = i
    features = np.load('Dataset/FaceFeatures/train/Neutral/'+file[:-3]+'npz')['a']
    faces_data.append(features.shape[0])
    print(i)
```

```python
x = len(faces_data)

for i in (pos_train):
    file = i
    features = np.load('Dataset/FaceFeatures/train/Positive/'+file[:-3]+'npz')['a']
    faces_data.append(features.shape[0])
    print(i)

y = len(faces_data)

print(y-x)
```

# Make List of Number of Faces in Val Images

```python
x = len(faces_data_val)

for i in (neg_val):
    file = i
    features = np.load('Dataset/FaceFeatures/val/Negative/'+file[:-3]+'npz')['a']
    faces_data_val.append(features.shape[0])
    print(i)

y = len(faces_data_val)

print(y-x)
```

```python
x = len(faces_data_val)

for i in (neu_val):
    file = i
    features = np.load('Dataset/FaceFeatures/val/Neutral/'+file[:-3]+'npz')['a']
    faces_data_val.append(features.shape[0])
    print(i)

y = len(faces_data_val)

print(y-x)
```

```python
x = len(faces_data_val)

for i in (pos_val):
    file = i
    features = np.load('Dataset/FaceFeatures/val/Positive/'+file[:-3]+'npz')['a']
    faces_data_val.append(features.shape[0])
    print(i)

y = len(faces_data_val)

print(y-x)
```
